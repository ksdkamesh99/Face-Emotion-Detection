```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```
dataset=pd.read_csv('Dataset/fer20131.csv')
dataset
```

```
emotion = {'Happy': 1,
           'Neutral': 0}
```

```
len(dataset)
```

```
images=[]
labels=[]
for i in range(0,len(dataset)):
  if dataset['emotion'][i]==3:
    d=dataset['pixels'][i].split(' ')
    o=np.array(d).astype('float32')
    o=o.reshape((48,48))
    o=cv2.cvtColor(o,cv2.COLOR_GRAY2RGB)
    o=o/255
    images.append(o)
    labels.append(1)
  if dataset['emotion'][i]==6:
    d=dataset['pixels'][i].split(' ')
    o=np.array(d).astype('float32')
    o=o.reshape((48,48))
    o=cv2.cvtColor(o,cv2.COLOR_GRAY2RGB)
    o=o/255
    images.append(o)
    labels.append(0)
```

```
images=np.array(images)
labels=np.array(labels)
```

```
np.save("images_smile_nuetral",images)
np.save("labels_smile_nuetral",labels)
```
